Loading data set

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/House Price India.csv")
```

Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df['Price'])
plt.show()
```

Bivariate Analysis

```python
sns.scatterplot(x='number of floors', y='Price', data=df)
plt.show()
```

Multivariate Analysis

```python
sns.pairplot(df[['Price', 'number of bedrooms', 'number of bathrooms']])
plt.show()
```

Descriptive statistics on the dataset.

```python
correlation_matrix = df[['Price', 'number of bedrooms', 'number of bathrooms']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
summary = df.describe()
print(summary)
```

Check missing value

```python
missing_values = df.isnull().sum()
print(missing_values)
```
